```python
import pandas as pd
import numpy as np
```

```python
subway1_8 = pd.read_csv('1_8호선_위치정보.csv', encoding='cp949')
subway1_8
```

```python
subway1_8 = subway1_8[['호선', '역명', '위도', '경도']]
subway1_8
```

```python
subway1_8['호선'] = subway1_8['호선'].astype('str')
```

```python
subway1_8.info()
```

```python
subway1_8.loc[:, '호선'] = subway1_8['호선'].str[:] + '호선'
```

```python
subway1_8.head(2)
```

```python
subway9 = pd.read_csv('9호선_위치정보.csv')
subway9.head(2)
```

```python
subway9 = subway9.dropna(axis=0)
```

```python
subway9.drop(subway9.columns[0], axis=1, inplace=True)
subway9.rename({'선명': '호선'}, axis=1, inplace=True)
subway9 = subway9[['호선', '역명', '위도', '경도']]
subway9.head(2)
```

```python
subway = pd.concat([subway1_8, subway9], axis=0)
subway
```

```python
subway.to_csv('1_9호선_위치정보.csv', encoding='utf-8', index=False)
```
