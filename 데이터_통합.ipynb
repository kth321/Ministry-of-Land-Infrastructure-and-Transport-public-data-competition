```python
logistic = pd.read_csv('물류량_가중치.csv')
logistic.head(2)
```

```python
JDY_addr = pd.read_csv('/content/drive/MyDrive/공모전/KTH/data/최종데이터/JDY_주소추가.csv')
JDY_addr.head(2)
```

```python
JDY_addr_IC = pd.merge(JDY_addr, ic, how='inner', on='차량기지')
JDY_addr_IC.info()
```

```python
JDY_addr_IC.rename({'역주소': '구분'}, axis=1, inplace=True)
JDY_addr_IC.head(2)
```

```python
JDY_addr_IC.info()
```

```python
JDY_addr_IC_logistic = pd.merge(JDY_addr_IC, logistic, how='inner', on='구분')
JDY_addr_IC_logistic.head(2)
```

```python
before_scale = JDY_addr_IC_logistic.iloc[:, 9:]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
JDY_addr_IC_logistic.iloc[:, 9:] = scaler.fit_transform(before_scale)
JDY_addr_IC_logistic
```

```python
JDY_addr_IC_logistic.to_csv('KTH_JDY_merge_final.csv', index=False, encoding='utf-8')
```
