```python
import pandas as pd
```

```python
subway = pd.read_csv('지하철역2.csv', encoding='cp949')
subway
```

```python
addresses = subway['구주소']
addresses
```

```python
new_addresses = []
for address in addresses:
  new_addresses.append(address.split()[1])
```

```python
subway['역주소'] = new_addresses
subway.drop(['구주소', '도로명주소', '전화번호'], axis=1, inplace=True)
subway
```
