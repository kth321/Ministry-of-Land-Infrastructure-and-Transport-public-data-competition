```python
import pandas as pd
import numpy as np
```

```python
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)
pd.options.display.float_format = '{:.6f}'.format
```

```python
ic_loc = pd.read_csv('IC_정보.csv', encoding='cp949')
ic_loc = ic_loc[['IC/JC명', 'X좌표값', 'Y좌표값']]
ic_loc.head(2)
```

```python
ic_loc.info()
```

```python
ic_loc = ic_loc[ic_loc['IC/JC명'].str[-2:] == 'IC']
ic_loc.head(2)
```

```python
ic_loc.rename({'IC/JC명': 'IC명', 'X좌표값': '경도', 'Y좌표값': '위도'}, axis=1,
              inplace=True)
ic_loc = ic_loc[['IC명', '위도', '경도']]
ic_loc.head(2)
```

```python
ic_loc.info()
```

```python
print(ic_loc[ic_loc['IC명'] == '여주IC'])
print(ic_loc[ic_loc['IC명'] == '남안성IC'])
```

```python
ic_loc = ic_loc[(ic_loc['경도'] <= 127.65425) & (ic_loc['위도'] >= 36.975677)]
ic_loc.reset_index(drop=True, inplace=True)
ic_loc
```
