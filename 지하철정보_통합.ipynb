```python
subway_info = pd.read_csv('지하철역정보.csv')
subway_info
```

```python
subway_info['역주소'] = subway_info['역주소'].str[6:]
subway_info
```

```python
data = json.load(open('지하철정보.json'))
info_df = pd.DataFrame(data['DATA'])
info_df
```

```python
info_df = info_df[['line_num', 'station_nm', 'station_cd']]
```

```python
info_df = info_df.rename({'line_num': '호선', 'station_nm': '역명',
      'station_cd': '역사코드'}, axis=1).sort_values('호선').reset_index(drop=True)
info_df
```

```python
merge_data = pd.merge(info_df, subway_info, on='역명',
                      how='inner')
merge_data
```
